```python
avgperp = []
for i in range(0,len(people)):
    if people[i]=='cd':
        text = tcd
    if people[i]=='sc':
        text = tsc
    if people[i]=='kh':
        text = tkh
    if people[i]=='am':
        text = tam
    if people[i]=='db':
        text = tdb
    if people[i]=='sk':
        text = tsk
    if people[i]=='bs':
        text = tbs
    if people[i]=='bis':
        text = tbis
    if people[i]=='jm':
        text = tjm
    if people[i]=='jsc':
        text = tjsc
    if people[i]=='ma':
        text = tma
    if people[i]=='ab':
        text = tab
    if people[i]=='hge':
        text = thge
    if people[i]=='wg':
        text = twg
    if people[i]=='am':
        text = tam
    if people[i]=='mk':
        text = tmk
    if people[i]=='gr':
        text = tgr
    if people[i]=='cds':
        text = tcds
    blob = t(text)
    polarity = []
    for sentence in blob.sentences:
        if sentence.sentiment.subjectivity>0.0:
            polarity.append(sentence.sentiment.polarity)
    print(polarity)
    avgperp.append(sum(polarity)/len(polarity)) if not len(polarity) == 0 else 0
    print(avgperp)

    bar = [i for i in range(len(polarity))]
    fig = plt.figure(figsize = (10,5))

    #Polarity pro Sentence für people[i]
    plt.bar(bar,polarity)
    plt.ylabel("Sentiment")
    plt.xlabel(range(0,len(blob.sentences)-1))
    plt.title("Polarity "+people[i])
    plt.show()

fig = plt.figure(figsize = (10,5))
# Durchschnittliche Polarity pro Person
plt.bar(bar,avgperp)
plt.ylabel("Durchschnitts-Sentiment")
plt.xlabel(range(0,len(people)-1))
plt.title("Average Polarity")
plt.show()
    
# avg = sum(polarity)/len(polarity) if not len(polarity) == 0 else 0
```
