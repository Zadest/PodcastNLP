```python
import json
import os
import sentimentAnalysis
import matplotlib.pyplot as plt
from textblob_de import TextBlobDE
```

```python
firstPodcast = None
mdr_podcasts = []
path = os.path.join("data","refined","mdr")

folder = os.listdir(path)
for filename in folder:
    filepath = os.path.join(path,filename)
    with open(os.path.join(path,filename), "r", encoding="utf-8") as f:
        podcast_dict = json.load(f)
        mdr_podcasts.append(podcast_dict)

kekulee_all = []
schumann_all = []
for podcast in mdr_podcasts:
    kekulee = []
    schumann = []
    for key in podcast:
        part = podcast[key]
        if part[0] == "Alexander KekulÃ©":
            kekulee.append(part[1])
            continue
        if part[0] == "Camillo Schumann":
            schumann.append(part[1])
    kekulee_all.append(kekulee)
    schumann_all.append(schumann)
```

```python
polarity_avg_kekulee = []
pos_polarity_avg_kekulee = []
neg_polarity_avg_kekulee = []
for i, folge in enumerate(kekulee_all):
    complete = "".join(folge)
    blob = TextBlobDE(complete)
    polarity = []
    pos_polarity = []
    neg_polarity = []
    for sentence in blob.sentences:
        polarity.append(sentence.sentiment.polarity)
        
        if sentence.sentiment.polarity >= 0:
            pos_polarity.append(sentence.sentiment.polarity)
        else:
            neg_polarity.append(sentence.sentiment.polarity)
    
    avg = sum(polarity)/len(polarity) if not len(polarity) == 0 else 0 
    pos_avg = sum(pos_polarity)/len(pos_polarity) if not len(pos_polarity) == 0 else 0 
    neg_avg = sum(neg_polarity)/len(neg_polarity) if not len(neg_polarity) == 0 else 0
    print(pos_avg,neg_avg)
    polarity_avg_kekulee.append(avg)
    pos_polarity_avg_kekulee.append(pos_avg)
    neg_polarity_avg_kekulee.append(neg_avg)
```

```python
polarity_avg_schuhmann = []
for i, folge in enumerate(schumann_all):
    complete = "".join(folge)
    blob = TextBlobDE(complete)
    polarity = []
    for sentence in blob.sentences:
        polarity.append(sentence.sentiment.polarity)
    avg = sum(polarity)/len(polarity) if not len(polarity) == 0 else 0 
    polarity_avg_schuhmann.append(avg)
polarity_avg_schuhmann
```

```python
fig = plt.figure(figsize=(12,8), dpi= 100, facecolor='w', edgecolor='k')
plt.bar(range(0,len(pos_polarity_avg_kekulee)),pos_polarity_avg_kekulee,label="KekulÃ©+")
plt.bar(range(0,len(neg_polarity_avg_kekulee)),neg_polarity_avg_kekulee,label="KekulÃ©-")
plt.ylabel("Avg Polarity")
plt.xlabel("Folge")
plt.legend()
plt.show()
```

```python
fig, axs = plt.subplots(1,2,figsize=(12,6),sharex='col', sharey='row')
fig.legend()
fig.suptitle("Sentiment Vergleich")

plt.ylabel("Avg Polarity")
plt.xlabel("Folge")

axs[0].bar(range(0,len(polarity_avg_kekulee)),polarity_avg_kekulee,label="KekulÃ©",color="orange")
axs[0].set_title("KekulÃ©")
axs[1].bar(range(0,len(polarity_avg_schuhmann)),polarity_avg_schuhmann,label="Schumann")
axs[1].set_title("Schumann")
```
